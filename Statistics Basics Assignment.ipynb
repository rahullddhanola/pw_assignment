```python
#Q1.Implement a Python program to compute the mean, median, and mode of a given list of numbers.

# Import the statistics module
import statistics

# Given list of numbers
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

# Calculate mean
mean_value = statistics.mean(numbers)

# Calculate median
median_value = statistics.median(numbers)

# Calculate mode
mode_value = statistics.mode(numbers)

# Display results
print(f"Numbers: {numbers}")
print(f"Mean: {mean_value}")
print(f"Median: {median_value}")
print(f"Mode: {mode_value}")
```

```python
#Q2.Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python:
# Import necessary modules
import statistics
import math

# Given datasets
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Calculate mean of each list
mean_x = statistics.mean(list_x)
mean_y = statistics.mean(list_y)

# Calculate covariance
covariance = sum((x - mean_x)*(y - mean_y) for x, y in zip(list_x, list_y)) / (len(list_x)-1)

# Calculate correlation coefficient
std_x = statistics.stdev(list_x)
std_y = statistics.stdev(list_y)
correlation = covariance / (std_x * std_y)

# Display results
print(f"Covariance: {covariance}")
print(f"Correlation Coefficient: {correlation}")
```

```python
#Q3.Write a Python script to draw a boxplot for the following numeric list andidentify its outliers.
# Import required libraries
import matplotlib.pyplot as plt
import numpy as np

# Given data
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Draw boxplot
plt.boxplot(data, vert=True, patch_artist=True, showmeans=True)
plt.title("Boxplot of the given data")
plt.ylabel("Values")
plt.show()

# Identify outliers using IQR method
Q1 = np.percentile(data, 25)  # First quartile
Q3 = np.percentile(data, 75)  # Third quartile
IQR = Q3 - Q1  # Interquartile Range

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = [x for x in data if x < lower_bound or x > upper_bound]

print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")
print(f"Outliers: {outliers}")
```

```python
#Q.Write Python code to compute the correlation between the two lists:
import statistics

# Given data
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Calculate means
mean_ad = statistics.mean(advertising_spend)
mean_sales = statistics.mean(daily_sales)

# Calculate standard deviations
std_ad = statistics.stdev(advertising_spend)
std_sales = statistics.stdev(daily_sales)

# Calculate correlation coefficient
correlation = sum((x - mean_ad)*(y - mean_sales) for x, y in zip(advertising_spend, daily_sales)) / ((len(advertising_spend)-1) * std_ad * std_sales)

print(f"Correlation coefficient: {correlation}")
```
