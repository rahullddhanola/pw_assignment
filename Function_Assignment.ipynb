```python
#Q2. Explain the concept of function arguments and parameters in Python?
'''In Python, parameters are variables listed in a function definition, while arguments are the actual values passed when calling the function.
Eample:
def greet(name):         # 'name' is a parameter
    print("Hello", name)

greet("Rahul")           # "Rahul" is an argument  '''
```

```python
#Q3. What are the different ways to define and call a function in Python?
'''In Python, functions can be defined and called in various ways depending on the requirement:
1.Regular Function :Defined using the def keyword with parameters. Called by passing arguments in order.
example-
def greet(name):
    print("Hello", name)
greet("Rahul")

2.Function with Default Arguments :You can assign default values to parameters. If no argument is passed, the default is used.
example-
def greet(name="Guest"):
    print("Hello", name)

greet()          # Uses default
greet("Rahul")   # Overrides default

3.Function with Keyword Arguments :Arguments can be passed using parameter names, allowing flexible ordering.
example-
def greet(name, message):
    print(f"{message}, {name}")

greet(name="Rahul", message="Good morning")

4.Variable-Length Arguments :Use *args for variable numbers of positional arguments and **kwargs for keyword arguments.
example-
def add(*numbers):
    print(sum(numbers))

add(1, 2, 3, 4)

5.Lambda Function :A short, anonymous function defined using the lambda keyword; used for simple operations.
example-
square = lambda x: x * x
print(square(5))  # Output: 25  '''



```

```python
#Q5.What are iterators in Python and how do they differ from iterables?
'''In Python, an iterable is any object you can loop over (like lists, strings), while an iterator is an object that produces 
items one at a time using next().
You get an iterator from an iterable by using the iter() function.
Example:
nums = [1, 2, 3]       # Iterable  
it = iter(nums)        # Iterator  
print(next(it))        # Output: 1  '''
```

```python
#Q6. Explain the concept of generators in Python and how they are defined?
'''A generator is a special type of function that returns an iterator and generates values one at a time using the yield keyword instead of return.
example:
def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1
* Generators save memory because they don't store all values at once.
* Each call to next() resumes from where yield paused.  '''
```

```python
#Q7. What are the advantages of using generators over regular functions?
'''Generators are more memory-efficient than regular functions because they yield values one at a time instead of storing the entire result.
They are ideal for working with large data or infinite sequences.
Generators also preserve state between calls, making them useful for lazy evaluation.
They simplify custom iterator creation using the yield keyword.
example:
def gen():
    yield 1
    yield 2

for i in gen():
    print(i)
    
Unlike a list-returning function, this uses minimal memory and produces values one at a time. '''
```

```python
#Q8.What is a lambda function in Python and when is it typically used?
'''A lambda function is a small, anonymous function defined using the lambda keyword.
It can take any number of arguments but has only one expression.
Lambda functions are used for short, anonymous functions typically in places like map(), filter(), or sorted().
They’re useful when a quick function is needed without formally defining it using def.
example-
square = lambda x: x * x
print(square(5))  # Output: 25
'''
```

```python
#Q9.Explain the purpose and usage of the `map()` function in Python ?
'''The map() function applies a given function to each item in an iterable (like a list) and returns a map object (which can be converted to a list).
example:
nums = [1, 2, 3]
squared = list(map(lambda x: x * x, nums))
print(squared)  # Output: [1, 4, 9]
'''
```

```python
#Q1.Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list?
def sum_even_num(n):
    even_sum = 0
    for i in n:
        if i%2==0:
            even_sum += i
    return even_sum
list1 = [12,34,31,20,47]
sum_even_num(list1)
```

```python
#Q3.Implement a Python function that takes a list of integers and returns a new list containing the squares of each number?
def sq_num(list):
    sq = []
    for i in list:
        sq.append(i**2)
    return sq

list1 = [1,2,3,4,5]
sq_num(list1)
```

```python
#Q4.Write a Python function that checks if a given number is prime or not from 1 to 200?
def is_prime(n):
    if n <= 1:
        return False  # 1 and numbers less than 1 are not prime
    for i in range(2, int(n**0.5) + 1):  # Check up to the square root of n
        if n % i == 0:
            return "Not a prime number"
    return "it is a prime number"

is_prime(8)
```

```python
#Q5.Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms?
def fib(n):
    a, b = 0, 1
    for i in range(n):
        yield a
        a, b = b, a + b

fib(10)
f = fib(10)
print(next(f))
print(next(f))
print(next(f))
```

```python
#Q6.Write a generator function in Python that yields the powers of 2 up to a given exponent?
def powers_of_two(max_exponent):
    for i in range(max_exponent + 1):
        yield 2 ** i
powers_of_two(5)
gen = powers_of_two(5)
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
```

```python
#Q7. Implement a generator function that reads a file line by line and yields each line as a string?
import requests

def read_url_lines(url):
    response = requests.get(url)
    lines = response.text.splitlines()
    for line in lines:
        yield line

url = 'https://example-files.online-convert.com/document/txt/example.txt'

for line in read_url_lines(url):
    print(line)

```

```python
#Q8.Use a lambda function in Python to sort a list of tuples based on the second element of each tuple?

data = [(1, 3), (4, 7), (5, 3), (6, 2)]

# Sort using lambda
sorted_data = sorted(data, key=lambda x: x[1])

print(sorted_data)
```

```python
#Q9.Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit?

celsius_temps = [20, 28, 34, 49, 72]

fahrenheit_temps = list(map(lambda x: (x * 9/5) + 32, celsius_temps))

print("Celsius:", celsius_temps)
print("Fahrenheit:", fahrenheit_temps)



```

```python
#Q10. Create a Python program that uses `filter()` to remove all the vowels from a given string?
def remove_vowels(s):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda char: char not in vowels, s))

input_str = "Hello, my name is rahul?"
result = remove_vowels(input_str)

print("Original String:", input_str)
print("Without Vowels:", result)
```

Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
product of the price per item and the quantity. The product should be increased by 10,- € if the value of the
order is smaller than 100,00 €.

Write a Python program using lambda and map

```python
orders = [
    [34587, 40.95, 4],
    [98762, 56.80, 5],
    [77226, 32.95, 3],
    [88112, 24.99, 3]
]

# Using map and lambda
result = list(map(lambda order: (
    order[0],
    order[1]*order[2] if order[1]*order[2] >= 100 else order[1]*order[2] + 10
), orders))

print(result)
```
