```python
#Question 1: Explain the differences between AI, ML, Deep Learning (DL), and Data Science (DS).
"""Answer: Artificial Intelligence (AI), Machine Learning (ML), Deep Learning (DL), and Data Science (DS) are closely related but distinct fields.
They often overlap but have unique definitions, purposes, and applications.
1. Artificial Intelligence (AI):
 * Definition: AI is a branch of computer science that aims to create systems capable of performing tasks that typically require human intelligence.
 * Key Goal: Simulate human intelligence in machines.
 * Techniques: Rule-based systems, expert systems, reasoning, planning, NLP, ML, robotics.
 * Example Applications: Virtual assistants (Siri, Alexa), chatbots, self-driving cars, game-playing AI.
 * Scope: AI is the broadest concept; ML and DL are subsets of AI.

2. Machine Learning (ML):
 * Definition: ML is a subset of AI that focuses on enabling machines to learn patterns from data and make decisions or predictions without explicit programming.
 * Key Goal: Train models on data so they can generalize and predict outcomes for unseen data.
 * Techniques: Supervised learning, unsupervised learning, reinforcement learning.
 * Example Applications: Email spam detection, product recommendations, fraud detection.
 * Relationship to AI: ML is one of the primary ways to achieve AI capabilities.

3. Deep Learning (DL):
 * Definition: DL is a subset of ML that uses neural networks with multiple layers (deep neural networks) to model complex patterns in large datasets.
 * Key Goal: Automatically extract features and learn representations from raw data.
 * Techniques: Convolutional Neural Networks (CNN), Recurrent Neural Networks (RNN), Transformers.
 * Example Applications: Image recognition, speech recognition, language translation, autonomous vehicles.
 * Relationship to ML: DL is an advanced form of ML, particularly useful for unstructured data like images, audio, and text.

4. Data Science (DS):
 * Definition: Data Science is an interdisciplinary field that combines statistics, mathematics, computer science, and domain knowledge to extract insights from data.
 * Key Goal: Collect, process, analyze, and interpret data to support decision-making.
 * Techniques: Data cleaning, visualization, statistical modeling, ML/DL models, predictive analytics.
 * Example Applications: Customer behavior analysis, business forecasting, healthcare analytics, recommendation systems.
 * Relationship to AI/ML: DS uses ML/DL as tools to analyze data but is broader, including data preparation, visualization, and business insights."""
```

```python
#Question 2: What are the types of machine learning? Describe each with one real-world example.
"""Answer:Machine Learning (ML) is broadly categorized into three main types: Supervised Learning, Unsupervised Learning, and Reinforcement Learning.
Each type has different goals, techniques, and applications.
1.Supervised Learning:
 * Definition: In supervised learning, the model is trained on a labeled dataset, meaning each input has a corresponding correct output. 
               The goal is to learn the mapping between inputs and outputs.
 * Goal: Predict outcomes for unseen data based on past examples.
 * Techniques: Regression (predicting continuous values), Classification (predicting discrete categories).
 * Example:
     * Email spam detection: The algorithm is trained with emails labeled as "spam" or "not spam" and learns to classify new incoming emails.

2.Unsupervised Learning:
 * Definition: In unsupervised learning, the model is trained on unlabeled data.The algorithm tries to find hidden patterns or structure in the 
               data without predefined labels.
 * Goal: Discover relationships, clusters, or reduce dimensionality in data.
 * Techniques: Clustering, Association Rule Learning, Dimensionality Reduction.
 * Example:
     * Customer segmentation: Retail companies cluster customers based on purchasing behavior to create targeted marketing campaigns.

3.Reinforcement Learning:
 * Definition: Reinforcement learning involves an agent learning to make decisions by interacting with an environment. The agent receives feedback
               in the form of rewards or penalties and improves its strategy over time.
 * Goal: Learn optimal actions to maximize cumulative reward.
 * Techniques: Q-learning, Deep Q-Networks, Policy Gradient methods.
 * Example:
     * Self-driving cars: The car learns to navigate traffic safely by receiving rewards for successful maneuvers and penalties for mistakes."""
```

```python
#Question 3: Define overfitting, underfitting, and the bias-variance tradeoff in machine learning.
"""Answer:Machine learning models aim to generalize well to unseen data. Overfitting, underfitting, and the bias-variance tradeoff are key concepts 
that describe how models perform and generalize.
1.Overfitting:
 * Definition: Overfitting occurs when a machine learning model learns the training data too well, including noise and minor fluctuations, 
               rather than the underlying patterns.
 * Characteristics:
    * High accuracy on training data.
    * Poor performance on unseen/test data.
    * Model is too complex (many parameters or deep networks for small datasets).
 * Example: A model predicting housing prices fits the training data perfectly but fails to predict prices for new houses accurately.
 * Solution: Use simpler models, regularization (L1/L2), pruning, or more training data.

2.Underfitting:
 * Definition: Underfitting occurs when a machine learning model is too simple to capture the underlying patterns in the data.
 * Characteristics:
    * Low accuracy on training and test data.
    * Model fails to learn from the data.
    * Insufficient features, low model complexity, or inadequate training.
 * Example: Using a linear model to fit data that has a quadratic relationship, resulting in poor predictions.
 * Solution: Increase model complexity, add relevant features, or train for longer.

3.Bias-Variance Tradeoff:
 * Definition: The bias-variance tradeoff describes the balance between bias (error due to oversimplification) and variance (error due to overfitting).
    * Bias: Error from making strong assumptions in the learning algorithm. High bias → underfitting.
    * Variance: Error from sensitivity to small fluctuations in the training data. High variance → overfitting.
 * Goal: Minimize total error by finding the right balance between bias and variance.
 * Example:
    * A simple linear regression may underfit (high bias).
    * A deep neural network with limited data may overfit (high variance).
    * Proper regularization and feature selection help achieve low bias and variance.

Conclusion:
 * Overfitting → model too complex, memorizes training data.
 * Underfitting → model too simple, fails to capture patterns.
 * Bias-Variance Tradeoff → balancing model complexity to achieve good generalization."""
```

```python
#Question 4: What are outliers in a dataset, and list three common techniques for handling them.
"""Answer:
1.Definition of Outliers:
 * Outliers are data points that significantly differ from other observations in a dataset.
 * They may result from measurement errors, data entry errors, or genuine extreme values.
 * Outliers can distort statistical analysis and machine learning models, affecting mean, variance, and predictions.
Example:
 * In a dataset of students’ test scores (0–100), a score of 200 would be an outlier.

2. Common Techniques for Handling Outliers:
(a) Removal
 * Simply delete the outlier data points from the dataset.
 * Best when outliers are due to errors and not representative of the population.
 * Example: Remove a negative age entry in a customer dataset.

(b) Transformation
 * Apply mathematical transformations (e.g., log, square root) to reduce the effect of outliers.
 * Helps stabilize variance and make data more normally distributed.
 * Example: Use log transformation for income data with extremely high values.

(c) Imputation or Capping
 * Replace extreme values with a threshold value or statistical measure like median or percentile.
 * Known as Winsorization when extreme values are capped.
 * Example: Cap salaries above the 95th percentile to the 95th percentile value.

Conclusion:
Outliers are extreme values that can skew analysis if not handled properly. Common strategies include removal, transformation, and 
imputation/capping, depending on the nature of the data and the problem."""
```

```python
#Question 5: Explain the process of handling missing values and mention one imputation technique for numerical and one for categorical data.
"""Answer:
1.Handling Missing Values in a Dataset
Missing values occur when some data entries are not recorded or lost, which can affect data analysis and machine learning model performance. 
Handling them is a crucial step in data preprocessing.

Process to handle missing values:
1.Identify missing values
 * Check which columns and rows contain missing data using techniques like isnull() in Python or summary statistics.
2.Analyze the cause and pattern
 * Determine whether missing values are random (MCAR) or systematic (MAR/MNAR).
 * This helps choose the appropriate handling method.
3.Decide on a strategy
 * Options include removing, imputing, or predicting missing values.
4.Implement the chosen method
 * Use suitable techniques for numerical or categorical data to fill in missing values.

2. Imputation Techniques:
(a) Numerical Data
   * Technique: Mean/Median Imputation
   * Replace missing numerical values with the mean or median of the column.
   * Example: If the age column has missing values, fill them with the median age of the dataset.

(b) Categorical Data
   * Technique: Mode Imputation
   * Replace missing categorical values with the most frequent category.
   * Example: If the gender column has missing values, fill them with the mode (e.g., “Male” if it occurs most often).

Conclusion:
Handling missing values is a key part of data preprocessing. For numerical data, mean or median imputation is common, while for categorical data, 
mode imputation is widely used. The choice of method depends on the data type, distribution, and missingness pattern."""
```

```python
#Question 6: Write a Python program that:
# ● Creates a synthetic imbalanced dataset with make_classification() from sklearn.datasets.
# ● Prints the class distribution

from sklearn.datasets import make_classification
from collections import Counter

#  Create a synthetic imbalanced dataset
X, y = make_classification(
    n_samples=1000,        # total number of samples
    n_features=10,         # number of features
    n_informative=5,       # number of informative features
    n_redundant=2,         # number of redundant features
    n_clusters_per_class=1,
    weights=[0.9, 0.1],    # imbalance: 90% of class 0, 10% of class 1
    flip_y=0,              # no label noise
    random_state=42
)

#  Print the class distribution
class_distribution = Counter(y)
print("Class distribution:", class_distribution)
```

```python
#Question 7: Implement one-hot encoding using pandas for the following list of colors:
#['Red', 'Green', 'Blue', 'Green', 'Red']. Print the resulting dataframe.

import pandas as pd

#  Create the list of colors
colors = ['Red', 'Green', 'Blue', 'Green', 'Red']

#  Convert the list into a DataFrame
df = pd.DataFrame(colors, columns=['Color'])

#  Apply one-hot encoding using get_dummies
one_hot_df = pd.get_dummies(df, columns=['Color']).astype(int)

#  Print the resulting DataFrame
print(one_hot_df)
```

```python
#Question 8: Write a Python script to:
#● Generate 1000 samples from a normal distribution.
#● Introduce 50 random missing values.
#● Fill missing values with the column mean.
#● Plot a histogram before and after imputation.


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Generate 1000 samples from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)
df = pd.DataFrame(data, columns=['Value'])

# Step 2: Introduce 50 random missing values
missing_indices = np.random.choice(df.index, size=50, replace=False)
df.loc[missing_indices, 'Value'] = np.nan

# Step 3: Plot histogram before imputation
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.hist(df['Value'].dropna(), bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram Before Imputation')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Step 4: Fill missing values with column mean
mean_value = df['Value'].mean()
df['Value'] = df['Value'].fillna(mean_value)  # Corrected

# Step 5: Plot histogram after imputation
plt.subplot(1,2,2)
plt.hist(df['Value'], bins=30, color='salmon', edgecolor='black')
plt.title('Histogram After Imputation')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

```

```python
#Question 9: Implement Min-Max scaling on the following list of numbers [2, 5, 10, 15,20] using sklearn.preprocessing.MinMaxScaler. 
#Print the scaled array.


from sklearn.preprocessing import MinMaxScaler
import numpy as np

# Step 1: Original list of numbers
numbers = [2, 5, 10, 15, 20]

# Step 2: Convert the list into a 2D NumPy array (required by MinMaxScaler)
numbers_array = np.array(numbers).reshape(-1, 1)

# Step 3: Initialize MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))

# Step 4: Fit and transform the data
scaled_numbers = scaler.fit_transform(numbers_array)

# Step 5: Print the scaled array
print("Scaled array:\n", scaled_numbers)
```

```python
#Question 10: You are working as a data scientist for a retail company. You receive a customer transaction dataset that contains:
# ● Missing ages,
# ● Outliers in transaction amount,
# ● A highly imbalanced target (fraud vs. non-fraud),
# ● Categorical variables like payment method.
# Explain the step-by-step data preparation plan you’d follow before training a machine learning model. 
# Include how you’d address missing data, outliers, imbalance, and encoding.

"""Before training a machine learning model, it is crucial to clean, preprocess, and transform the data to ensure the model performs well. 
Here’s a step-by-step plan for the given dataset:

Step 1: Understand and Explore the Dataset
 * Check basic information: Use .info() and .describe() in Python to understand data types, missing values, and summary statistics.
 * Visualize distributions: Histograms, boxplots, and scatterplots can help identify outliers, skewness, and patterns.

Step 2: Handle Missing Values
* Column with missing ages:
  * If the missing values are numerical (like age), use imputation.
  * Option 1: Mean/Median Imputation: Replace missing ages with the median (robust to outliers).
  * Option 2: Predictive Imputation: Use other features (like transaction amount, location, or customer segment) to predict missing ages.
Note: Avoid dropping rows unless a large portion of data is missing.

Step 3: Handle Outliers
* Transaction amount outliers:
  * Detect outliers using IQR method (Q1 - 1.5*IQR and Q3 + 1.5*IQR) or Z-score method.
  * Options to handle:
       * Capping (Winsorization): Replace extreme values with percentile values (e.g., 1st and 99th percentile).
       * Transformation: Apply log or square root to reduce skewness.
       * Remove: Only if clearly erroneous and small in number.

Step 4: Handle Imbalanced Target
* Fraud vs. non-fraud: Highly imbalanced classes can bias the model toward majority class.
* Techniques to address imbalance:
  * Resampling:
       * Oversampling minority class: e.g., SMOTE (Synthetic Minority Oversampling Technique).
       * Undersampling majority class: Randomly reduce non-fraud samples.
  * Class weighting: Use algorithms that allow weighting classes (e.g., class_weight='balanced' in scikit-learn).
  * Evaluation metrics: Use metrics like precision, recall, F1-score, ROC-AUC instead of accuracy.

Step 5: Encode Categorical Variables
* Payment method or other categorical variables:
  * Nominal categories: Use One-Hot Encoding (pd.get_dummies) to avoid ordinality.
  * Ordinal categories: Use Label Encoding if categories have an inherent order.
  * High cardinality: Consider target encoding or embedding for categories with many unique values.

Step 6: Feature Scaling
* Scale numerical features (like transaction amount, age) for algorithms sensitive to feature magnitude (e.g., logistic regression, KNN, SVM).
* Techniques:
  * Min-Max Scaling for normalization.
  * Standardization (Z-score scaling) for mean=0, std=1.

Step 7: Split Data
* Split dataset into training and test sets (e.g., 70%-30%) before resampling to prevent data leakage.

Step 8: Optional Feature Engineering
* Create new features: e.g., transaction frequency, average purchase amount.
* Handle multicollinearity: Remove highly correlated features to avoid redundancy.

Step 9: Ready for Modeling
* After preprocessing:
  * No missing values.
  * Outliers handled.
  * Categorical variables encoded.
  * Target balanced (or class weights applied).
  * Features scaled if necessary.

The data is now clean and suitable for training machine learning models like logistic regression, decision trees, random forests, or gradient boosting.

Summary Table

| Issue                 | Solution                                                                  |
| --------------------- | ------------------------------------------------------------------------- |
| Missing ages          | Impute with median or predictive modeling                                 |
| Outliers in amounts   | Detect with IQR/Z-score; cap, transform, or remove                        |
| Imbalanced target     | Oversample minority, undersample majority, or use class weights           |
| Categorical variables | One-hot encoding (nominal), label/target encoding (ordinal/high-cardinal) |
| Feature scaling       | Min-Max or standardization for numeric features                           |"""
```
